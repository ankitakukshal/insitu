## Construction of Stress Polygons

Normal Faulting Polygon

```
Sv_x = np.arange(0, 30010, 10)
Sv_y = Sv_x

plt.figure(figsize=(10,7))
p1 = plt.plot(Sv_x, Sv_y, color='green')
p2 = plt.plot(Sv, Sv, 'o', color='black')

# For normal faulting (NF), lower limit of Shmin
Sh = 0.6 * Sv

"three points of NF triangle, Sh < SH < Sv"
# define Sh lower bound, Sh upper bound, SH lower bound, SH upper bound
Sh_low = Sh
Sh_up = Sv
SH_low = Sh
SH_up = Sv

# point1: Sh_low, Sh_low
point1_x = Sh_low; point1_y = SH_low
# point2: Sh_low, SH_up
point2_x = Sh_low; point2_y = SH_up
# point3: Sh_up, SH_up
point3_x = Sh_up; point3_y = SH_up

po1 = plt.plot(point1_x, point1_y, 'o', color='blue')
po2 = plt.plot(point2_x, point2_y, 'o', color='red')
po3 = plt.plot(point3_x, point3_y, 'o', color='orange')

# create polygon
poly_x = np.array([point1_x, point2_x, point3_x, point1_x])
poly_y = np.array([point1_y, point2_y, point3_y, point1_y])
poly = plt.plot(poly_x, poly_y, color='purple')

plt.legend((p1[0], p2[0], po1[0], po2[0], po3[0], poly[0]), (['Sv line', 'Sv data = 11000 psi', 'point(Sh low, Sh low)', 'point(Sh low, SH up)', 'point(Sh up, SH up)', 'RF Polygon']))
plt.title('Stress Polygon for Normal Faulting Regime (Sh < SH < Sv)')
plt.xlabel('Shmin (psi)'); plt.ylabel('SHmax (psi)')
plt.xlim(0, 30000); plt.ylim(0, 30000)
```

## Wellbore breakout and Tensile fracture line on Stress polygons

#### **Tensile failure**

Circumferential stress equation (omitting cooling stress $\sigma^{\Delta T}$):

$$Sh+SH-2(SH-Sh)-2P_p-(P_m-P_p)=T_0$$

Derived $SH$ as a function of $Sh$ becomes:

$$SH=3Sh - (T_0+2P_p+(P_m-P_p))$$

#### **Wellbore breakout**

Circumferential stress equation (omitting cooling stress $\sigma^{\Delta T}$):

$$Sh+SH-2(SH-Sh)\cos(\pi-W_{bo})-2P_p-(P_m-P_p)=C_0$$

Derived $SH$ as a function of $Sh$ becomes:

$$SH=\frac{C_0+2P_p+(P_m-P_p)-Sh(1+2\cos(\pi-W_{bo}))}{1-2\cos(\pi-W_{bo})}$$

Where $(\pi-W_{bo})$ is in radian, or $(180^o-W_{bo})$ in degree.

```
# Input data for polygons
Pp = 4400 # psi
Sv = 11000

# Input data for tensile and breakout lines
Pm = 5900 # mud pressure
C0 = 14000 # unconfined compressive strength, UCS
Wbo = 0 # degree, wellbore breakout width

# lower limit of Shmin, from Sv
Sh = 0.6 * Sv

# upper limit of SHmax, from Sv and Pp
SH = (3.1 * (Sv - Pp)) + Pp

# axes of plot
Sv_x = np.arange(0, (SH + 10000), 10)
Sv_y = Sv_x

plt.figure(figsize=(10,10))

# plot Sv line and Sv data
p1 = plt.plot(Sv_x, Sv_y, color='green')
p2 = plt.plot(Sv, Sv, 'o', color='black')

# for Normal Faulting (NF) regime
nf_x = np.array([Sh, Sh, Sv, Sh])
nf_y = np.array([Sh, Sv, Sv, Sh])
nf = plt.plot(nf_x, nf_y, color='blue')

# for Strike Slip (SS) regime
ss_x = np.array([Sh, Sv, Sv, Sh])
ss_y = np.array([Sv, Sv, SH, Sv])
ss = plt.plot(ss_x, ss_y, color='red')

# for Reverse Faulting (RF) regime
rf_x = np.array([Sv, Sv, SH, Sv])
rf_y = np.array([Sv, SH, SH, Sv])
rf = plt.plot(rf_x, rf_y, color='black')

"Tensile fracture line"
Sh_line = Sv_x
SH_tensile = (3 * Sh_line) - ((2 * Pp) + (Pm - Pp))
tensile = plt.plot(Sh_line, SH_tensile, '--')

"Wellbore breakout line"
SH_breakout = (C0 + (2 * Pp) + (Pm - Pp) - (Sh_line *  (1 + 2 * np.cos(np.deg2rad(180 - Wbo))))) / (1 - 2 * np.cos(np.deg2rad(180 - Wbo)))
breakout = plt.plot(Sh_line, SH_breakout, '--')

plt.legend((p1[0], p2[0], nf[0], ss[0], rf[0], tensile[0], breakout[0]), 
           ['Sv line', 'Sv data', 'Normal Faulting (NF) Polygon', 
            'Strike Slip (SS) Polygon', 'Reverse Faulting (RF) Polygon',
            'Tensile fracture line', 'Wellbore breakout line'])

plt.title('Stress Polygons (Zoback-o-gram)')
plt.xlabel('Shmin (psi)'); plt.ylabel('SHmax (psi)')
plt.xlim(0, SH+10000); plt.ylim(0, SH+10000)
plt.gca().set_aspect('equal')
```

## Homework question. Find the 5 points intersection of tensile and breakout lines with the stress polygon

![stress-polygon](https://user-images.githubusercontent.com/51282928/77507238-9b6f4100-6e9a-11ea-9a87-5b4b0ebf36ae.PNG)

### Point 1. Intersection NF polygon & breakout line

#### Minimum stress at intersection

is the lower bound of Shmin

```
Sh1 = 0.6 * Sv
print("Minimum horizontal stress at point 1:", Sh1, "psi")
```

#### Maximum stress at intersection

substitute `Sh1` to SH equation of breakout line

```
SH1 = (C0 + (2 * Pp) + (Pm - Pp) - (Sh1 *  (1 + 2 * np.cos(np.deg2rad(180 - Wbo))))) / (1 - 2 * np.cos(np.deg2rad(180 - Wbo)))
print("Maximum horizontal stress at point 1:", SH1, "psi")
```

### Point 2. Intersection RF polygon & breakout line

Intersection between **SH-Sh normal line** and breakout line

The first equation is SH-Sh normal line (in implicit format): 

$$SH-Sh=0$$

The second equation is the breakout line (in implicit format):

$$\frac{C_0+2P_p+(P_m-P_p)-Sh(1+2\cos(\pi-W_{bo}))}{1-2\cos(\pi-W_{bo})}-SH=0$$

$SH$ and $Sh$ can be simultaneously solved (2-variable linear solver)

```
# Solving Sh and SH simultaneously

from scipy.optimize import fsolve

def f(y):
  Sh, SH = y
  # normal line
  f1 = SH - Sh
  # breakout line
  f2 = ((C0 + (2 * Pp) + (Pm - Pp) - (Sh *  (1 + 2 * np.cos(np.deg2rad(180 - Wbo))))) / (1 - 2 * np.cos(np.deg2rad(180 - Wbo)))) - SH
  return[f1, f2]

solve = fsolve(f, [1, 1]) # initial guess
```

#### Minimum stress at intersection

```
Sh2 = solve[0]
print("Minimum stress at point 2:", Sh2, "psi")
```

#### Maximum stress at intersection

```
SH2 = solve[1]
print("Maximum stress at point 2:", SH2, "psi")
```

#### Maximum stress at intersection

substitute `Sh3` to SH equation of tensile line

```
SH3 = (3 * Sh3) - ((2 * Pp) + (Pm - Pp))
print("Maximum horizontal stress at point 3:", SH3, "psi")
```

### Point 4. Intersection RF polygon & tensile line

**Better to start from computing maximum stress (SH) at intersection first**

#### Maximum stress at intersection

is the upper bound of $SH$ in RF regime

```
SH4 = (3.1 * (Sv - Pp)) + Pp
print("Maximum horizontal stress at point 4:", SH4, "psi")
```

#### Minimum stress at intersection

substitute `SH4` to tensile line equation, becomes:

$$Sh=\frac{SH + (T_0+2P_p+(P_m-P_p))}{3}$$

```
Sh4 = (SH4 + ((2 * Pp) + (Pm - Pp))) / 3
print("Minimum horizontal stress at point 4:", Sh4, "psi")
```

### Point 5. Intersection breakout line & tensile line in NF regime

Intersection between tensile line and breakout line

The first equation is tensile line (in implicit format): 

$$3Sh-(T_0+2P_p+(P_m-P_p))-SH=0$$

The second equation is the breakout line (in implicit format):

$$\frac{C_0+2P_p+(P_m-P_p)-Sh(1+2\cos(\pi-W_{bo}))}{1-2\cos(\pi-W_{bo})}-SH=0$$

$SH$ and $Sh$ can be simultaneously solved (2-variable linear solver)

```
# Solving Sh and SH simultaneously

from scipy.optimize import fsolve

def f(y):
  Sh, SH = y
  # tensile line
  f1 = (3 * Sh) - ((2 * Pp) + (Pm - Pp)) - SH
  # breakout line
  f2 = ((C0 + (2 * Pp) + (Pm - Pp) - (Sh *  (1 + 2 * np.cos(np.deg2rad(180 - Wbo))))) / (1 - 2 * np.cos(np.deg2rad(180 - Wbo)))) - SH
  return[f1, f2]

solve = fsolve(f, [1, 1]) # initial guess
```

#### Minimum stress at intersection

```
Sh5 = solve[0]
print("Minimum horizontal stress at point 5:", Sh5, "psi")
```

#### Maximum stress at intersection

```
SH5 = solve[1]
print("Maximum horizontal stress at point 5:", SH5, "psi")
```
